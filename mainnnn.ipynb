```python
%%time
i = 1
for i in range(1,101):
    try:
        print(f'iteración {i}')
        url = f'https://www.futbin.com/22/players?page={i}'
        data_1 = data1(url)
        dicfinal['name_player'] = dicfinal['name_player']+data_1['name_player']
        dicfinal['media'] = dicfinal['media']+data_1['media']
        dicfinal['posicion'] = dicfinal['posicion']+data_1['posicion']
        dicfinal['tipo'] = dicfinal['tipo']+data_1['tipo']
        dicfinal['precio'] = dicfinal['precio']+data_1['precio']
        dicfinal['skill'] = dicfinal['skill']+data_1['skill']
        dicfinal['wf'] = dicfinal['wf']+data_1['wf']
        dicfinal['ritmo'] = dicfinal['ritmo']+data_1['ritmo']
        dicfinal['shot'] = dicfinal['shot']+data_1['shot']
        dicfinal['pas'] = dicfinal['pas']+data_1['pas']
        dicfinal['drib'] = dicfinal['drib']+data_1['drib']
        dicfinal['def'] = dicfinal['def']+data_1['def']
        dicfinal['phy'] = dicfinal['phy']+data_1['phy']
        dicfinal['popularity'] = dicfinal['popularity']+data_1['popularity']
        dicfinal['base_stats'] = dicfinal['base_stats']+data_1['base_stats']
        dicfinal['igs'] = dicfinal['igs']+data_1['igs']
        dicfinal['plataforma'] = dicfinal['plataforma']+data_1['plataforma']
        dicfinal['time'] = dicfinal['time']+data_1['time']

        dicfinal['nacionalidad'] = dicfinal['nacionalidad']+data_1['nacionalidad']
        dicfinal['club'] = dicfinal['club']+data_1['club']
        dicfinal['liga'] = dicfinal['liga']+data_1['liga']
        #dicfinal['cara'] = dicfinal['cara']+data_1['cara']
        
        print('finalizado')
        i = i + 1
    except:
        pass
    
    if i == 800:
        break
with open('data3.json', 'w') as fp:
    json.dump(dicfinal, fp)
```
